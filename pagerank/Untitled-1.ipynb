```python
import pandas as pd
import networkx as nx

df = pd.read_csv('OnlineRetail.csv', encoding = 'unicode-escape')
```

```python
df[100:121]
```

```python
df.head()
```

```python
df1 = df.sample(30)
```

```python
G = nx.from_pandas_edgelist(df1, source='CustomerID', target='StockCode', create_using=nx.DiGraph())
```

```python
pos = nx.bipartite_layout(G, nodes=df1['CustomerID'], scale= 0.5)
```

```python
nx.draw(G, pos, with_labels=True)
```
